```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import Perceptron, LogisticRegression
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, random_state=42)
```

```python
perceptron = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
perceptron.fit(X, y)
```

```python
logistic_regression = LogisticRegression(random_state=42)
logistic_regression.fit(X, y)
```

```python
def plot_decision_boundary(model, X, fill=False, alpha=1.0):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    if fill:
        plt.contourf(xx, yy, Z, alpha=alpha, cmap='coolwarm')
    else:
        plt.contour(xx, yy, Z, alpha=alpha, colors='k', linestyles='--')
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Perceptron")
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
plot_decision_boundary(perceptron, X, fill=True, alpha=0.2)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.subplot(1, 2, 2)
plt.title("Logistic Regression")
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
plot_decision_boundary(logistic_regression, X, fill=True, alpha=0.2)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.tight_layout()
plt.show()
```
